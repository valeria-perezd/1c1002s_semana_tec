# Actividad de Graficación

```python
import numpy as np
import pandas as pd
```

```python
#Archivo de excel
df = pd.read_excel("/content/Sales Data Presentation - Dashboards.xlsx",
                    sheet_name="Raw Data")
print(df.head())
```

```python
#Gráfica de barras con los 5 productos menos vendidos
df_quantity = df.groupby('ProdName')['Quantity'].sum().sort_values(ascending=True).head(5)
print(df_quantity)
df_quantity.plot(kind='bar', color='yellow')
```

```python
#Gráfica de caja para la columna de Quantity del producto con nombre Scented Olive Oil Lotion Gift Basket
df_olive_oil = df[df['ProdName'] == 'Scented Olive Oil Lotion Gift Basket']['Quantity']
print(df_olive_oil)
df_olive_oil.plot(kind='box',patch_artist=True)
```
