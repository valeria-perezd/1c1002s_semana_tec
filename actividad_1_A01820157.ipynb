# Actividad de Numpy

```python
#Importar numpy como primer paso
import numpy as np
```

```python
#Arreglo de 10 cincos
arreglo_cincos =np.full(10, 5)
print(arreglo_cincos)
```

```python
#Arreglo de enteros pares desde 10, inclusive, hasta 50, inclusive
arreglo_de_enteros = np.arange (10,51,10)
print (arreglo_de_enteros)
```

```python
#Matriz cuadrada de 4 por 4 de números enteros aleatorios
matriz_aleatoria=np.random.randint(0, 100, size=(4, 4))
print(matriz_aleatoria)
```

```python
#Arreglo de 10 números aleatorios entre 0 y 1 y calcula la suma de todos sus elementos
elementos=np.random.rand(10)
print(elementos)
print("suma")
print(np.sum(elementos))
```

```python
#Genera un arreglo de 10 números aleatorios entre 0 y 1 y calcula su media
elementos=np.random.rand(10)
print(elementos)
print("media")
print(np.mean(elementos))
```
