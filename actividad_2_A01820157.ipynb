# Actividad de Pandas

```python
#importamos Numpy y Pandas
import numpy as np
import pandas as pd
```

```python
#dataframe ficticio de ventas de frutas
df = pd.DataFrame({'Product': ['Carrots', 'Broccoli', 'Banana', 'Banana',
 'Beans', 'Orange', 'Broccoli', 'Banana'],
 'Category': ['Vegetable', 'Vegetable', 'Fruit', 'Fruit',
 'Vegetable', 'Fruit', 'Vegetable', 'Fruit'],
 'Quantity': [8, 5, 3, 4, 5, 9, 11, 8],
 'Amount': [270, 239, 617, 384, 626, 610, 62, 90]})
print(df)
```

```python
#Selecciona e imprime  las columnas de Product y Quantity
print (df[["Product", "Quantity"]])
```

```python
# Selecciona e imprime los renglones con índices 4 y 6
print(df.loc[[4, 6]])
```

```python
#Usando el método groupby, agrupa usando la columna Product y muestra la suma de la columna Quantity
print(df.groupby("Product")["Quantity"].sum())
```

```python
#Usando el método groupby, agrupa usando la columna Categoryy muestra la suma de la columna Amount
print(df.groupby("Category")["Amount"].sum())
```
